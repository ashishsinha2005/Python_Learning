```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data\\WAC.csv")
data.head()
```

```python
#1. Write a Pandas program to display the dimensions or shape of the World alcohol consumption dataset.
#Also extract the column names from the dataset.
df=data.copy()
print('dimensions of the dataset',df.shape)
print('column names from the dataset',df.columns)
```

```python
#2. Write a Pandas program to select first 2 rows, 2 columns and specific two columns
#from World alcohol consumption dataset
print('select first 2 rows\n',df.head(2))
print('select first 2 columns\n',df.iloc[:,:2])
print('specific two columns\n',df[['Year','Country']])
```

```python
# 3. Write a Pandas program to select random number of rows,
# fraction of random rows from World alcohol consumption dataset.
print('select random number of rows\n',df.sample(5))
print('fraction of random rows\n',df.sample(frac=0.2))
```

```python
# 4. Write a Pandas program to find and drop the missing
# values from World alcohol consumption dataset
print('find missing values\n',df.isna().sum())
print('another way to display the rows having missing value\n',df[df.isnull().any(axis=1)])
print('dropping missing values\n',df.dropna(inplace=True))
print(df.isna().sum())
```

```python
# 5. Write a Pandas program to remove the duplicates
# from 'WHO region' column of World alcohol consumption dataset.
print("check duplicate values in 'WHO region' column")
print(df['WHO region'].duplicated().sum())
print("remove duplicate values in 'WHO region' column")
df.drop_duplicates(subset='WHO region',inplace=True)
df['WHO region'].duplicated().sum()
```

```python
# 6. Write a Pandas program to find out the alcohol consumption of a given
# year from the world alcohol consumption dataset.
df=data.copy()
df.head()
print('alcohol consumption of a given year 1989')
df
df[df['Year']==1989].head(10)
```

```python
# 7. Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from
# the world alcohol consumption dataset.
print('alcohol consumption details in the year 1987 or 1989')
df[(df['Year']==1987)|(df['Year']==1989)].head(10)
```

```python
# 8. Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985'
# from the world alcohol consumption dataset.
print("alcohol consumption details by the 'Americas' in the year '1989'")
df[(df['WHO region']=='Americas')&(df['Year']==1989)]
```

```python
# 9. Write a Pandas program to find out the alcohol consumption details in the year '1986'
# where WHO region is 'Western Pacific' and country is 'VietNam'
# from the world alcohol consumption dataset.
print("alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 'Viet Nam'" )
df[(df['WHO region']=='Western Pacific')&(df['Country']=='Viet Nam')&(df['Year']==1986)]
```

```python
# 10. Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989'
#where WHO region is 'Americas'
# from the world alcohol consumption dataset.
print("alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas'")
df_temp=df[(df['Year']==1986)|(df['Year']==1989)]
df_temp[df_temp['WHO region']=='Americas']
```

```python
# 11. Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989'
# where WHO region is 'Americas' or 'Europe'
# from the world alcohol consumption dataset.
print("the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe'")
df_temp=df[(df['Year']==1986)|(df['Year']==1989)]
df_temp[(df_temp['WHO region']=='Americas')|(df_temp['WHO region']=='Europe')]
```

```python
# 13. Write a Pandas program to find out the records where consumption of beverages per person average >=5
# and Beverage Types is Beer from world alcohol consumption dataset.
print("consumption of beverages per person average >=5 and Beverage Types is Beer")
df[(df['Display Value']>=5)&(df['Beverage Types']=='Beer')]
```

```python
# 14. Write a Pandas program to find out the records where consumption of beverages per person average >=4 and
# Beverage Types is Beer, Wine, Spirits from world alcohol consumption dataset.
print("consumption of beverages per person average >=4 and Beverage Types is Beer, Wine, Spirits")
df[(df['Display Value']>=4)&((df['Beverage Types']=='Beer')|(df['Beverage Types']=='Wine')|(df['Beverage Types']))]
```

```python
# 15. Write a Pandas program to filter the specified columns and records by range from world alcohol consumption dataset.
print("specified columns and records by range")
cols_to_keep=['Year','Country','Beverage Types','Display Value']
df_temp=df[cols_to_keep]
min_range=4
max_range=8
df_temp[(df_temp['Display Value']>=min_range)&(df_temp['Display Value']<=max_range)]
```

```python
# 16. Write a Pandas program to filter those records where WHO region contains
# "Ea" substring from world alcohol consumption dataset.
print('WHO region contains "Ea" substring')
df1=df[df['WHO region'].apply(lambda x:'Ea' in x)]
df1
#alternative way df[df['WHO region'].str.contains("Ea")]
```

```python
# 17. Write a Pandas program to filter those records where WHO region matches with multiple
# values (Africa, Eastern Mediterranean, Europe)
#from world alcohol consumption dataset.
print("WHO region matches with multiple values (Africa, Eastern Mediterranean, Europe)")
df[(df['WHO region']=='Africa')|(df['WHO region']=='Mediterranean')|(df['WHO region']=='Europe')].head()
```

```python
# 18. Write a Pandas program to filter those records
# which not appears in a given list from world alcohol consumption dataset.
print("records which not appears in a given list")
exclude_list = ['Africa', 'Europe']
df[~df['WHO region'].isin(exclude_list)].head()
```

```python
# 19. Write a Pandas program to filter all records where the average consumption of beverages per person
# from .5 to 2.50 in world alcohol consumption dataset.
print("average consumption of beverages per person from .5 to 2.50")
df[(df['Display Value']>=0.5)&(df['Display Value']<=2.5)].head()
```

```python
# 20. Write a Pandas program to find average consumption of wine per person greater than 2
# in world alcohol consumption dataset
print("average consumption of wine per person greater than 2")
df[(df['Display Value']>2)&(df['Beverage Types']=='Wine')]
```

```python
# 21. Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30
# from world alcohol consumption dataset
print("rows based on row numbers ended with 0, like 0, 10, 20, 30")
filtered_rows=[rows for rows in range(df.shape[0]) if rows%10==0]
df.loc[filtered_rows,:]
#alternative way filter(regex='0$', axis=0)
```

```python
# 22. Write a Pandas program to select consecutive columns and also
#select rows with Index label 0 to 9 with some columns from
# world alcohol consumption dataset.
print("select rows with Index label 0 to 9 with some columns")
print(df.loc[:9,['Country','Year','Beverage Types']])
print("select consecutive columns")
print(df.loc[:9,:])
```

```python
# 23. Write a Pandas program to rename all and only some of the column names from world alcohol consumption dataset.
print("rename all columns")
df1=df.copy()
df1.rename(columns={'Year':'Year1','WHO region':'WHO region1','Country':'Country 1','Beverage Types':'Beverage Types1',
'Display Value':'Display Value1'},inplace=True)
print(df1.columns)
print("rename some of the column names")
df1.rename(columns={'Year1':'Year2','Country 1':'Country 2'},inplace=True)
print(df1.columns)
```
