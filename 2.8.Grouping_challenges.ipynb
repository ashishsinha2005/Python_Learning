```python
# 1. Write a Pandas program to split the following dataframe
# into groups based on school code. Also check the type of GroupBy object.
student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004'],
'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
'age': [12, 12, 13, 13, 14, 12],
'height': [173, 192, 186, 167, 151, 159],
'weight': [35, 32, 33, 30, 31, 32],
'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
print("Original Dataframe")
print(student_data)
print("split the following dataframe into groups based on school code")
result=student_data.groupby(by='school_code')
for name,group in result:
    print(name)
    print(group)
print("check the type of GroupBy object\n",type(result))
```

```python
# 2. Write a Pandas program to split the following dataframe by school code
# and get mean, min, and max value of age for each school.
student_data = pd.DataFrame({
'school_code': ['s001','s002','s003','s001','s002','s004'],
'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
'age': [12, 12, 13, 13, 14, 12],
'height': [173, 192, 186, 167, 151, 159],
'weight': [35, 32, 33, 30, 31, 32],
'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
                            index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
print("Original Dataframe")
print(student_data)
result=student_data.groupby('school_code')
print("split the following dataframe by school code and get mean, min, and max value of age for each school")
for name, group in result:
    print('school code:',name)
    print('mean age\n',group['age'].mean())
    print('minimum age\n',min(group['age']))
    print('maximum age\n',max(group['age']))
#alternative way: student_data.groupby('school_code').agg({'age': ['mean', 'min', 'max']})
```
